# Q4 :

Write a program in C# Sharp to separate odd and even integers in separate arrays.
Go to the editor
Test Data :
Input the number of elements to be stored in the array :5
Input 5 elements in the array :
element - 0 : 25
element - 1 : 47
element - 2 : 42
element - 3 : 56
element - 4 : 32
Expected Output:
The Even elements are:
42 56 32
The Odd elements are :
25 47

```
using System;
public class Exercise10
{
    public static void Main()
    {
        int[] arr1 = new int[10];
        int[] arr2 = new int[10];
        int[] arr3 = new int[10];
        int M, D = 0, P = 0, n;


        Console.Write("Give the number of items :");
        n = Convert.ToInt32(Console.ReadLine());

        Console.Write("Input {0} items :\n", n);
        for (M = 0; M < n; M++)
        {
            Console.Write("element - {0} : ", M);
            arr1[M] = Convert.ToInt32(Console.ReadLine());
        }

        for (M = 0; M < n; M++)
        {
            if (arr1[M] % 2 == 0)
            {
                arr2[D] = arr1[M];
                D++;
            }
            else
            {
                arr3[P] = arr1[M];
                P++;
            }
        }

        Console.Write("\n Even items are : \n");
        for (M = 0; M < D; M++)
        {
            Console.Write("{0} ", arr2[M]);
        }

        Console.Write("\n Odd items are :\n");
        for (M = 0; M < P; M++)
        {
            Console.Write("{0} ", arr3[M]);
        }
        Console.Write("\n\n");
    }
}
```
